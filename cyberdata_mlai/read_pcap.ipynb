```python
file_name = "data/pcaps/2024-09-04-traffic-analysis-exercise.pcap"
```

```python
from scapy.all import rdpcap

packets = rdpcap(file_name)

for packet in packets:
    if packet.haslayer('TCP') and packet.dport == 80: # packet.haslayer('HTTP') does not find any packets
        if packet.haslayer('Raw'):
            print(packet['Raw'].show())
```

```python
import threading
import pyshark
import queue


def capture_packets(file_name, packet_queue):
    capture = pyshark.FileCapture(file_name)
    for packet in capture:
        if "HTTP" in packet:
            packet_queue.put(packet)  # Add the packet to the queue


# Create a queue to hold the packets
packet_queue = queue.Queue()

# Run the capture in a separate thread
thread = threading.Thread(target=capture_packets, args=(file_name, packet_queue))
thread.start()

# Wait for the thread to finish
thread.join()

# Retrieve packets from the queue
captured_packets = []
while not packet_queue.empty():
    captured_packets.append(packet_queue.get())

# Now you can use captured_packets as needed
for packet in captured_packets:
    print(packet.http)
```
